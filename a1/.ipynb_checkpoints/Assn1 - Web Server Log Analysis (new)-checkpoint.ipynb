
Below is an example formatting string and a short description of each element:
    
    format = r'%h %l %u %t \"%r\" %>s %b'
    
where

    %h          - remote host (ie the client IP)
    %l          - identity of the user determined by identd (not usually used since not reliable)
    %u          - user name determined by HTTP authentication
    %t          - time the server finished processing the request.
    %r          - request line from the client. ("GET / HTTP/1.0")
    %>s         - status code sent from the server to the client (200, 404 etc.)
    %b          - size of the response to the client (in bytes)
    

Remember we want the data to be a *list of dictionaries* as shown below:

    data = [
    {'%>s': '200',
     '%b': '6245',
     '%h': '199.72.81.55',
     '%l': '-',
     '%r': 'GET /history/apollo/ HTTP/1.0',
     '%t': '[01/Jul/1995:00:00:01 -0400]',
     '%u': '-'},
    {'%>s': '200',
     '%b': '3985',
     '%h': 'unicomp6.unicomp.net',
     '%l': '-',
     '%r': 'GET /shuttle/countdown/ HTTP/1.0',
     '%t': '[01/Jul/1995:00:00:06 -0400]',
     '%u': '-'},

````python
# Fill code here

path = "data/NASA_access_log_Jul95"
def openFile(path):
    data = []
    line_parser = apache_log_parser.make_parser("%h %l %u %t \"%r\" %>s %b")
    with codecs.open(path,encoding='utf-8', errors='ignore') as f:
        for line in f:
            try:
                log_line_data = line_parser(line)
                data.append(log_line_data)
            except:
                pprint(line)
            
    return data

data = openFile(path)
````

````python
#test = data[:2]

def cleanData(data):
    cleanup_dic = []
    for dic in data:
        #pprint(dic)
        small_dic = {}
        small_dic['%>s'] = dic['status']
        small_dic['%b'] = dic['response_bytes_clf']
        small_dic['%h'] = dic['remote_host']
        small_dic['%l'] = dic['remote_logname']
        small_dic['%r'] = dic['request_first_line']
        small_dic['%t'] = dic['time_received'].strip('[]')
        small_dic['%u'] = dic['remote_user']
        cleanup_dic.append(small_dic)
    return cleanup_dic
#dic['status'] 
#dic['response_bytes_clf']
#dic['remote_host']
#dic['remote_logname']
#dic['request_first_line']
#dic['time_received']
#dic['remote_user']
cleanedData = cleanData(data)
pprint(len(cleanedData))
#pprint(cleanedData)
````

Some of the rows in the *b* column contain '-' string, so we can't convert them with *astype*. We can apply a custom function to this column, that will convert all dashes to NaN, and the rest to floats, additionally converting from bytes to megabytes:

````python
def dash2nan(x):
    if x == '-':
        x = np.nan
    else:
        x = float(x)/1048576.
    return x
````

Apply the above fuction to `b` column of `df`. Try using `.apply()`

````python
# Fill code here
df["b"] = df["b"].apply(dash2nan)
df.head()
````

Now our data frame is ready!

## Part 4 : Server status codes

I would like to know how different server responses are distributed through time. Create a new Data Frame that counts the number of requests per status in every 2 hour time spans.

```
HTTP Status Code
----------------
200 - OK
302 - Found
304 - Not Modified
400 - Bad Request
403 - Forbidden 
404 - Not Found
500 - Internal Server Error
501 - Not Implemented
```

````python
# Fill code here
df1 = df.copy()
# df1['b'] = df1['b']*1000000
# df1.head()
````

````python
df1.tail()
````

````python

start = pd.to_datetime("1995-07-01 04:00:00")
end = pd.to_datetime("1995-07-28 18:00:00")
status = {'Bad Request': 400, 'Forbidden': 403, "Found":302, "Internal Server Error":500,"Not Found": 400, "Not Implemented":501, "Not Modified": 304, "OK":200}

listofDic = []
while (start < end):
    newDic = {}
    newDic['Time'] = start
    next2Hour = start + pd.to_timedelta(2,unit='h')
    #print(type(df1.index))
    indexing = (df1.index >= start) & (df1.index <= next2Hour) 
    
    counting = df[indexing].groupby('Status')['Status'].count()
    for k,v in status.items():
        try:
            newDic[k] = counting[v]
        except:
            newDic[k] = np.NAN
    listofDic.append(newDic) 
    start =  next2Hour 


````

````python
status_df = pd.DataFrame(listofDic)
status_df.index = status_df["Time"]
# status_df.head()
````

````python
#status_df.drop('Time', axis=1)
status_df = status_df.drop('Time', 1)
status_df.head()
````

````python
# Fill code here
df1 = df.copy()
start = pd.to_datetime("1995-07-01 00:00:00")
end = pd.to_datetime("1995-07-29 00:00:00")
status = {'Bad Request': 400, 'Forbidden': 403, "Found":302, "Internal Server Error":500,"Not Found": 400, "Not Implemented":501, "Not Modified": 304, "OK":200}

listofDic = []
while (start < end):
    newDic = {}
    newDic['Time'] = start
    nextDay = start + pd.to_timedelta(1,unit='d')
    indexing = (df1.index >= start) & (df1.index <= nextDay) 
    
    counting = df[indexing].groupby('Status')['Status'].count()
    for k,v in status.items():
        try:
            newDic[k] = counting[v]
        except:
            newDic[k] = np.NAN
    listofDic.append(newDic) 
    start =  nextDay 


````

One of the challenges for this part is that some entries are not the IP addresses but hostnames. Here is excerpt from pygeoip API. You can use these function to get the country code.

```
country_code_by_addr(addr) -- Returns 2-letter country code (e.g. US) from IP address.
    Parameters:	addr – IP address (e.g. 203.0.113.30)

country_code_by_name(hostname) -- Returns 2-letter country code (e.g. US) from hostname.
    Parameters:	hostname – Hostname (e.g. example.com)
```
